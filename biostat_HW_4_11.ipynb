### 6. Construct the likelihood function and find the maximum likelihood estimate (MLE) of the parameter

### 7. Construct from the prior and likelihood the posterior distribution (Gaussian)
$$p(z)=\frac{1}{\sqrt{2\pi}\bar \sigma}\exp{(-\frac{(z-\bar \mu)^2}{2\bar \sigma^2})}$$
$$\bar \mu = \frac{\frac{1}{\sigma_0^2} \mu_0 + \frac{N}{\sigma}\bar y}{\frac{1}{\sigma_0^2} + \frac{n}{\sigma^2}} \\
{\bar \sigma}^2 = \frac{1}{\frac{1}{\sigma_0^2} + \frac{n}{\sigma^2}}$$
$$\bar \mu =1.98,~~\bar \sigma = 0.04$$
